The top 10 universities that have higher instutitional expenditure per student, Havard Univeristy has more full-time students, around 7000.
```python
college.sort_values(by='Expend', ascending=0)[[ 'GradRate']].iloc[0:10].plot(subplots = True, 
                                                                                            figsize=(20,20), kind='bar')
```

Out of the top 10 universities that have higher instutitional expenditure per student, only Antioch University has a graduation rate of less than 80%.

The bottom 10 universities that have lower instutitional expenditure per student diverge less than the top 10.

```python
college.sort_values(by='Expend', ascending=1)[['FUndergrad', 'PUndergrad']].iloc[0:10].plot(subplots = True,fontsize=15, 
                                                                                            figsize=(20,20), kind='bar')
```

The bottom 10 universities that have lowest instutitional expenditure per student, Jamestown College has more full-time students (about 10000).
